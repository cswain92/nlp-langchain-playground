```python
## Count Lines, Words, Characters

def count_text_file(file_path):
    line_count = 0
    word_count = 0
    char_count = 0

    with open(file_path, 'r') as file:
        for line in file:
            line_count += 1
            word_count += len(line.split())
            char_count += len(line)

    return line_count, word_count, char_count


filepath = "../../data/sample.txt"
print(count_text_file(filepath))
```

```python
## Email Validator
import re

def is_valid_email(email):
    pattern = r'^[a-zA-Z0-9._+-]+@[a-zA-Z0-9-]+\.[a-zA-Z0-9-.]+$'
    return re.match(pattern, email) is not None


print(is_valid_email("test@example.com"))
print(is_valid_email("invalid-email"))
```

```python
## Word Frequency
def count_word_frequency(file_path):
    word_count = {}

    with open(file_path, 'r') as file:
        for line in file:
            words = line.split()

            for word in words:
                word = word.lower().strip('.,!?;:"')
                word_count[word] = word_count.get(word, 0) + 1

    return word_count


filepath = "../../data/sample.txt"
print(count_word_frequency(filepath))
```

```python
## Relative vs Absolute Path
relative_path = "subdir/file.txt"
absolute_path = os.path.abspath(relative_path)
print(f' Relative path: {relative_path}')
print(f' Absolute path: {absolute_path}')

## what is relative path and absolute path
# Relative Path: A relative path specifies the location of a file or directory in relation to the current working directory. It does not start with a root directory and is often used for files that are located within the same project or directory structure.
# Absolute Path: An absolute path specifies the complete path to a file or directory, starting from the root directory. It provides the full location of the file regardless of the current working directory. Absolute paths are often used when referencing files that are located outside of the current project or directory structure.
```

```python
## Path exist check
path = "sample.txt"
if os.path.exists(path):
    print(f' The path "{path}" exists.')
else:
    print(f' The path "{path}" does not exist.')
```

```python
## Check if a path is a file or directory
import os
path = "sample.txt"
if os.path.isfile(path):
    print(f' "{path}" is a file.')
elif os.path.isdir(path):
    print(f' "{path}" is a directory.')
else:
    print(f'The Path "{path}" does not exist.')
```
