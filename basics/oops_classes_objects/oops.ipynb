## 2. Inheritance

👉 Child gets features from parent

```python
class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal):
    pass

d = Dog()
d.speak()
```

####  The above code demonstrates inheritance in Python. The `Dog` class inherits from the `Animal` class and can use the `speak` method defined in the `Animal` class without needing to override it. The `Cat` class, on the other hand, overrides the `speak` method to provide a different implementation.

🧠 Think: Dog IS-A Animal

## 3. Polymorphism

👉 Same function, different behavior

#### Polymorphism is the ability of different classes to be treated as instances of the same class through inheritance. It allows methods to be used interchangeably, even if they belong to different classes. This is achieved through method overriding, where a subclass provides a specific implementation of a method that is already defined in its superclass.

```python
class Cat(Animal):
    def speak(self):
        print("Cat meows")

c = Cat()
c.speak()
```

#### The above code demonstrates inheritance and method overriding in Python. The `Animal` class has a method `speak()`, which is inherited by the `Dog` class without modification, while the `Cat` class overrides the `speak()` method to provide a different implementation.

```python
class Shape:
    def area(self):
        pass

class Circle(Shape):
    def __init__(self,radius):
        self.radius = radius

    def area(self):
        import math
        return math.pi * self.radius ** 2
    
circle = Circle(5)
print(circle.area())

```

#### The above is an example of polymorphism where the area method is defined in the base class Shape and overridden in the Circle class to provide a specific implementation for calculating the area of a circle.

## 4. Encapsulation
👉 Hide data (protect it)

```python
class Bank:
    def __init__(self):
        self.__balance = 1000  # private

    def get_balance(self):
        return self.__balance

b = Bank()
print(b.get_balance())  # 1000
# print(b.__balance)  # AttributeError: 'Bank' object has no attribute '__balance'


try:
    print(b.__balance)
except AttributeError as e:
    print(e)
finally:
    print("Accessing private attribute failed as expected.")
```

#### This above will work because we are accessing the private variable through a public method. However, if we try to access the private variable directly, it will raise an AttributeError.

```python
class BankNew:
    def __init__(self):
        self.__balance = 0 ## Private variable

    def deposit(self, amount):
        self.__balance += amount
        print(f"Deposited {amount}. Current balance: {self.__balance}")

    def withdraw(self, amount):
        if amount > self.__balance:
            print("Insufficient funds")
        else:
            self.__balance -= amount
            print(f"Withdrew {amount}. Current balance: {self.__balance}")

    def get_balance(self):
        return self.__balance

bank = BankNew()
bank.deposit(100)
bank.withdraw(30)
print(bank.get_balance())
# print(bank.__balance)  # Accessing private variable directly (not recommended)
```

#### the above example demonstrates encapsulation by keeping the balance private and providing public methods to interact with it.

## 5. Abstraction

👉 Show only important things

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        pass

class Car(Vehicle):
    def start_engine(self):
        print("Car engine started")

car = Car()
car.start_engine()
```

#### This above example demonstrates the use of an abstract base class (Vehicle) with an abstract method (start_engine). The Car class inherits from Vehicle and provides an implementation for the start_engine method.

## 6. Magic Methods

👉 Special methods (__something__)

#### Magic methods (also called dunder methods = double underscore) are special methods in Python that let you define how your objects behave with built-in operations.

####  Example: __init__ , __str__ ,  __add__

#### 🔹 Why they are useful

#### They let your custom class act like built-in types.

👉 Without magic methods → your object is “plain”
👉 With magic methods → your object behaves like +, print(), len(), etc.

```python
class Test:
    def __init__(self):
        print("Object created")

t = Test()
```

```python
class Test:
    def __init__(self):
        print("Object created")

    def __del__(self):
        print("Object destroyed")

t = Test()
```

```python
class Person:
    def __init__(self, name):
        self.name = name

p = Person("John")
print(p.name)
```

#### __str__ (Readable output)

#### 👉 Without __str__ → weird output like <__main__.Person object at 0x10ab370e0>
#### 👉 With it → clean output

```python
class Person:
    def __init__(self, name):
        self.name = name

    def __str__(self):
        return f"Person name is {self.name}"

p = Person("John")
print(p)
```

```python
class MyList:
    def __init__(self, items):
        self.items = items

    def __len__(self):
        return len(self.items)

obj = MyList([1, 2, 3])
print(len(obj))   # 3
```

```python
class Number:
    def __init__(self, value):
        self.value = value

    def __add__(self, other):
        return self.value + other.value

n1 = Number(10)
n2 = Number(20)

print(n1 + n2)   # 30
```

```python
class Person:
    def __init__(self, age):
        self.age = age

    def __eq__(self, other):
        return self.age == other.age

p1 = Person(25)
p2 = Person(25)

print(p1 == p2)   # True
```

```python
class Person:
    def __repr__(self):
        return "This is object representation"
    
p = Person()
print(p)   # This is object representation
```

## 7. Operator Overloading

👉 Customize operators like +

```python
class Add:
    def __init__(self, x):
        self.x = x

    def __add__(self, other):
        return self.x + other.x

a = Add(10)
b = Add(20)

print(a + b)  # 30
```

```python
class Animal:
    def __init__(self, name):
        self.name = name
        self.__type = "Living Being"   # 🔒 private variable

    def speak(self):
        print("Animal makes sound")

    def get_type(self):   # access private
        return self.__type


class Dog(Animal):   # ✅ inheritance
    def speak(self):  # ✅ polymorphism
        print(f"{self.name} says Bark")


class Cat(Animal):   # ✅ inheritance
    def speak(self):  # ✅ polymorphism
        print(f"{self.name} says Meow")


# Objects
d = Dog("Tommy")
c = Cat("Kitty")

d.speak()
c.speak()

print(d.get_type())
```
