#### Exception handling in Python is how you handle runtime errors gracefully so your program doesn’t crash.

### Basic Idea

#### Instead of your program stopping on an error, you catch it and decide what to do.


```python
try:
    x = int(input("Enter a number: "))
    print(10 / x)
except:
    print("Something went wrong!")
```

```python
try:
    x = int(input("Enter number: "))
    result = 10 / x
except ValueError:
    print("Invalid input! Please enter a number.")
except ZeroDivisionError:
    print("Cannot divide by zero!")

# print(10/0)
```

#### This above code will catch specific exceptions and provide more informative error messages to the user.

```python
try:
    x = int(input("Enter number: "))
except ValueError:
    print("Wrong input")
else:
    print("You entered:", x)
```

```python
try:
    file = open("test.txt")
except FileNotFoundError:
    print("File not found")
finally:
    print("This will always run")
```

```python
try:
    x = int(input("Enter number: "))
    print(10 / x)
except ValueError:
    print("Invalid input")
except ZeroDivisionError:
    print("Divide by zero error")
else:
    print("Success!")
finally:
    print("Execution completed")
```

```python
def check_age(age):
    if age < 18:
        raise Exception("Not eligible")

try:
    check_age(15)
except Exception as e:
    print(e)
```
