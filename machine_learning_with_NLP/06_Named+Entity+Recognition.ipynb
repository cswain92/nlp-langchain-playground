# 🧠 What is Named Entity Recognition (NER)?

👉 **NER = finding and classifying important real-world entities in text**

---

## 📌 Simple Definition

👉 *NER identifies names like people, places, organizations, dates, etc. from text.*

---

# 🧪 Example

```text
"Elon Musk founded SpaceX in the USA in 2002"
```

👉 Output:

* Elon Musk → PERSON
* SpaceX → ORG
* USA → GPE (Location)
* 2002 → DATE

---

# 🏷️ Common NER Labels (Must Know)

| Label   | Meaning                 | Example   |
| ------- | ----------------------- | --------- |
| PERSON  | Person name             | Elon Musk |
| ORG     | Organization            | Google    |
| GPE     | Location (country/city) | India     |
| DATE    | Date                    | 2024      |
| TIME    | Time                    | 10 AM     |
| MONEY   | Money                   | $100      |
| PERCENT | Percentage              | 50%       |

---

# 💻 Using spaCy (Best for NER)

```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = "Elon Musk founded SpaceX in the USA in 2002"

doc = nlp(text)

for ent in doc.ents:
    print(ent.text, "→", ent.label_)
```

---

# 🎯 Output


# ⚠️ Important Notes

* NER uses **context + ML models**
* Not just rules
* Better with larger models (`en_core_web_md`, `lg`)

````python
## Using spaCy (Best for NER)

import spacy

nlp = spacy.load("en_core_web_sm")

# text = "Elon Musk founded SpaceX in the USA in 2002"
text = "Virat Kohli plays for India"

doc = nlp(text)

for ent in doc.ents:
    print(ent.text, "→", ent.label_)
````


### 🧠 Why This Happened

You’re using:

en_core_web_sm

👉 This is a small model, so:

- Less training data
- Lower accuracy
- Sometimes confuses PERSON with ORG
