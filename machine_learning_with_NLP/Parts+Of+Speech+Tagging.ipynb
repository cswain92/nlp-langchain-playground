## 🧠 What is Part of Speech (POS) Tagging?

👉 POS tagging = label each word with its role in a sentence

Example:

"I am running fast"

👉 Tagged:

I/PRP am/VBP running/VBG fast/RB

# 🏷️ Part of Speech (POS) Tags – Cheat Sheet

## 🔗 Conjunctions & Determiners
- **CC** → Coordinating conjunction (and, but, or)  
- **CD** → Cardinal digit (1, 2, 100)  
- **DT** → Determiner (the, a, an)  
- **PDT** → Predeterminer (all, both)  

---

## 📍 Special Words
- **EX** → Existential “there” (there is, there are)  
- **FW** → Foreign word  
- **LS** → List marker (1), a), i))  
- **UH** → Interjection (oh!, wow!, err)  

---

## 🔗 Prepositions & Particles
- **IN** → Preposition/Subordinating conjunction (in, on, because)  
- **TO** → “to” (to go, to the store)  
- **RP** → Particle (give up, take off)  

---

## 📝 Nouns
- **NN** → Noun, singular (desk)  
- **NNS** → Noun, plural (desks)  
- **NNP** → Proper noun, singular (Harrison)  
- **NNPS** → Proper noun, plural (Americans)  

---

## 👤 Pronouns
- **PRP** → Personal pronoun (I, he, she)  
- **PRP$** → Possessive pronoun (my, his, hers)  
- **WP** → Wh-pronoun (who, what)  
- **WP$** → Possessive wh-pronoun (whose)  

---

## 🎨 Adjectives
- **JJ** → Adjective (big)  
- **JJR** → Comparative adjective (bigger)  
- **JJS** → Superlative adjective (biggest)  

---

## ⚡ Adverbs
- **RB** → Adverb (very, silently)  
- **RBR** → Comparative adverb (better)  
- **RBS** → Superlative adverb (best)  
- **WRB** → Wh-adverb (where, when)  

---

## ⚙️ Verbs
- **VB** → Base form (take)  
- **VBD** → Past tense (took)  
- **VBG** → Gerund/present participle (taking)  
- **VBN** → Past participle (taken)  
- **VBP** → Present (non-3rd person) (take)  
- **VBZ** → Present (3rd person) (takes)  
- **MD** → Modal (can, could, will)  

---

## 🔍 Others
- **POS** → Possessive ending (’s)  
- **WDT** → Wh-determiner (which)  

---

## 🧠 Quick Memory Tip

- **NN** → Noun (things)  
- **VB** → Verb (action)  
- **JJ** → Adjective (description)  
- **RB** → Adverb (how)

```python
from nltk.tokenize import word_tokenize
```

```python
from nltk import pos_tag
```

```python
text = "I am learning NLP and running fast"

words = word_tokenize(text)

tagged = pos_tag(words)

print(tagged)
```
