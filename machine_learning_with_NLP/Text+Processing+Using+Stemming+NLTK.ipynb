
## 🌱 Stemming (Easy Definition)

**Stemming is the process of reducing a word to its base or root form by removing suffixes or prefixes.**

👉 It may not always produce a real dictionary word.

---

## 🧠 Simple Understanding

Think of it like **cutting endings of words**:

| Original Word | Stem    |
| ------------- | ------- |
| running       | run     |
| playing       | play    |
| studies       | studi ❌ |

👉 Notice:

* “studi” is **not a real word** → that’s okay in stemming

---

## ⚡ Why We Use It

* To treat similar words as the same
* Reduce data size
* Improve model performance

Example:

```
run, running, runs → all become "run"
```

---

## ⚠️ Important Point

👉 Stemming is **fast but not always accurate**

---

## 🔁 Stemming vs Lemmatization (Quick Clarity)

* **Stemming** → cuts words (may be wrong)
* **Lemmatization** → returns correct dictionary word

Example:

```
studies → studi (stemming ❌)
studies → study (lemmatization ✅)
```

---

## 💻 Example in Python using NLTK

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

words = ["running", "playing", "studies"]

for word in words:
    print(word, "→", stemmer.stem(word))
```

## Porter Stemmer (Most Common)

👉 Default & widely used

Fast
Simple rules
Sometimes incorrect

````python
from nltk.stem import PorterStemmer
````

````python
stemming = PorterStemmer()
````

````python
for word in words:
    print(word+"--->"+stemming.stem(word))
````

## Snowball Stemmer (Better Version of Porter)

👉 Improved version of Porter

Supports multiple languages
More consistent

````python
from nltk.stem import SnowballStemmer
````

````python
snowballStemmer = SnowballStemmer('english')
````

````python
for word in words:
    print(word+"--->"+snowballStemmer.stem(word))
````

## Lancaster Stemmer (Aggressive ⚠️)

👉 Cuts words very aggressively

Faster than Porter
Can over-cut words

````python
from nltk.stem import LancasterStemmer
````

````python
lancasterStemmer = LancasterStemmer()
````

````python
for word in words:
    print(word+"--->"+lancasterStemmer.stem(word))
````

## Regexp Stemmer (Custom Rules)

👉 You define your own pattern

It basically takes a single regular expression and removes any prefix or suffix that matches the expression

from nltk.stem import RegexpStemmer

stemmer = RegexpStemmer('ing$|s$|e$')
stemmer.stem("playing")  # play

````python
from nltk.stem import RegexpStemmer
````

````python
reg_stemmer = RegexpStemmer('ing$|s$|e$|able$')
````

````python
reg_stemmer.stem('eating')
````

````python
reg_stemmer.stem('stable')
````
