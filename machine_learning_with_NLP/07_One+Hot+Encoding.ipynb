# 🔢 One Hot Encoding (OHE) in NLP

We will understand One Hot Encoding using 3 documents:

- D1: The Food is Good  
- D2: The Food is Bad  
- D3: Pizza is Amazing  

```python
documents = [
    "The Food is Good",
    "The Food is Bad",
    "Pizza is Amazing"
]
```

## 🔡 Step 1: Preprocessing

- Convert to lowercase
- Tokenize words

```python
from nltk.tokenize import word_tokenize

processed_docs = [word_tokenize(doc.lower()) for doc in documents]
processed_docs
```

## 📚 Step 2: Vocabulary

Vocabulary = Unique words from all documents

```python
vocab = set()

for doc in processed_docs:
    vocab.update(doc)

vocab = sorted(list(vocab))
vocab
```

## 🔢 Step 3: One Hot Encoding

Each word is represented as a binary vector:
- One position = 1  
- Rest = 0  

```python
import numpy as np

word_to_index = {word:idx for idx, word in enumerate(vocab)}
word_to_index
```

```python
def one_hot(word, vocab_size):
    vector = np.zeros(vocab_size, dtype=int)
    # print(vector)
    if word in word_to_index:
        vector[word_to_index[word]] = 1
    return vector
```

## 🧪 Step 4: Apply Encoding

```python
encoded_docs = []

for doc in processed_docs:
    encoded_doc = [one_hot(word, len(vocab)) for word in doc]
    # print(f"vector: {encoded_doc}\n")
    encoded_docs.append(encoded_doc)

encoded_docs
```

## 🔍 Understanding

Each word becomes a vector.

Example:
- food → [1,0,0,...]
- pizza → [0,1,0,...]

Each word lies on a separate axis (no relationship between them).

## ✅ Advantages

1. Easy to implement using Python  
   (sklearn OneHotEncoder, pandas get_dummies)

2. Simple and fast  
   (no training required)

## ❌ Disadvantages (Very Important)
### 1️⃣ Sparse Matrix → Overfitting

👉 Most values are 0

Example:

[0,0,0,0,1,0,0]

💡 With large vocab (50k words) → huge sparse matrix

### 2️⃣ Fixed Input Size

👉 Vocabulary must be predefined
👉 New word → problem

### 3️⃣ No Semantic Meaning ❌
good ≠ amazing (but meaning is similar)

Vectors:

good    → [0,0,0,1,0,0,0]
amazing → [0,0,0,0,0,0,1]

👉 No similarity captured

### 4️⃣ Out Of Vocabulary (OOV)

👉 New word:

"delicious"

❌ Not in vocab → cannot represent

🧠 Memory Trick (From Your Image)

👉 Think:

Each word = separate axis
No connection between axes
⚡ Final Understanding
Word → Unique ID (vector)
Sentence → collection of vectors
🎯 Interview One-Liner

👉 One-Hot Encoding represents each word as a binary vector with one active position, but it fails to capture semantic meaning and leads to sparse representations.

## 🧠 Key Takeaway

- One Hot Encoding assigns a unique vector to each word  
- It does not capture meaning or relationships  
- Leads to sparse and high-dimensional data
