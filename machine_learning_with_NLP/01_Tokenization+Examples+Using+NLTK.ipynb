# 🧠 NLP Basics: Tokenization, Corpus, Vocabulary

Great—you’re asking the **right foundation question**.  
If you understand this well, everything in NLP becomes easier.

---

## 🧠 Big Picture

👉 NLP works in a hierarchy:

```

Corpus → Documents → Sentences → Words → Tokens

```

---

## 📚 1️⃣ What is a Corpus?

👉 A **corpus** is a collection of multiple documents.

### Example:
```

Document 1: "I love NLP"
Document 2: "NLP is amazing"
Document 3: "I love coding"

```

👉 All together = **Corpus**

---

## 📄 2️⃣ What is a Document?

👉 A **document** is a single piece of text.

Examples:
- One email  
- One tweet  
- One paragraph  

```

"I love NLP" → Document

```

---

## 🔡 3️⃣ What is Vocabulary?

👉 **Vocabulary = Unique words from the corpus**

### Example:
```

"I love NLP"
"NLP is amazing"
"I love coding"

```

👉 Vocabulary:
```

["I", "love", "NLP", "is", "amazing", "coding"]

```

💡 No duplicates allowed.

---

## 🔤 4️⃣ Words vs Tokens

- **Words** → Raw text  
- **Tokens** → Split/processed units  

---

## ✂️ 5️⃣ What is Tokenization?

👉 Tokenization is the process of breaking text into smaller units (tokens).

---

### 🧪 Example 1: Word Tokenization

```

Text: "I love NLP"

```

👉 Tokens:
```

["I", "love", "NLP"]

```

---

### 🧪 Example 2: Sentence Tokenization

```

Text: "I love NLP. It is powerful."

```

👉 Tokens:
```

["I love NLP.", "It is powerful."]

```

---

### 🧪 Example 3: With Punctuation

```

Text: "Hello, world!"

```

👉 Tokens:
```

["Hello", ",", "world", "!"]

```

---

## 🔥 Full Flow Example

```

Corpus:
↓
["I love NLP", "NLP is amazing"]

Documents:
↓
"I love NLP"
"NLP is amazing"

Vocabulary:
↓
["I", "love", "NLP", "is", "amazing"]

Tokenization:
↓
["I", "love", "NLP"]

````

## 💻 Python Practice

### Using NLTK

```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize

text = "I love NLP. It is amazing!"

sentences = sent_tokenize(text)
words = word_tokenize(text)

print("Sentences:", sentences)
print("Words:", words)
````

---

### Using spaCy

```python
import spacy

nlp = spacy.load("en_core_web_sm")

doc = nlp("I love NLP. It is amazing!")

for token in doc:
    print(token.text)
```

* Vocabulary = Unique words in library
* Tokens = Words split from book
