## What are Stopwords?

👉 Stopwords are common words that don’t add much meaning to a sentence

Examples:

is, am, are, the, a, an, in, on, at, and, to

### 💡 Simple Definition

👉 Stopwords are frequently used words that are usually removed from text to improve NLP processing.

🎯 Why Remove Stopwords?
- Reduce data size
- Improve model performance
- Focus on meaningful words

```python
from nltk.corpus import stopwords
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
stop_words = set(stopwords.words('english'))
```

```python
stop_words
```

```python
words = word_tokenize(paragraph)
```

```python
words
```

```python
filtered = [word for word in words if word.lower() not in stop_words]
```

```python
filtered
```

```python
sentences = sent_tokenize(paragraph)
```

```python
sentences
```

```python
from nltk.stem import PorterStemmer
```

```python
stemmer = PorterStemmer()
```

```python
## Apply Stopwords And Filter And then Apply Stemming(PorterStemmer)

for i in range(len(sentences)):
     words = word_tokenize(sentences[i])
     words = [ stemmer.stem(word) for word in words if word.lower() not in stop_words]
     sentences[i]=' '.join(words)# converting all the list of words into sentences


```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
from nltk import pos_tag
```

```python
  >>> import nltk
  >>> nltk.download('averaged_perceptron_tagger_eng')
```

```python
def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return 'a'
    elif tag.startswith('V'):
        return 'v'
    elif tag.startswith('N'):
        return 'n'
    elif tag.startswith('R'):
        return 'r'
    else:
        return 'n'
```

```python
## Apply Stopwords And Filter And then Apply Lemmatization

for i in range(len(sentences)):
     words = word_tokenize(sentences[i])
     tagged_words = pos_tag(words)
     print(tagged_words)
     words = [ lemmatizer.lemmatize(word, get_wordnet_pos(tag)) for word, tag in tagged_words if word.lower() not in stop_words]
     sentences[i]=' '.join(words)# converting all the list of words into sentences
```
