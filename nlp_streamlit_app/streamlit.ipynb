# 🚀 Introduction to Streamlit

**Streamlit** is an open-source Python framework used to build **interactive web applications** quickly and easily. It is widely used in **data science, machine learning, and NLP projects** because it allows developers to create user interfaces using only Python—without needing HTML, CSS, or JavaScript.

---

## 🧠 How It Works

Streamlit converts Python scripts into web apps.  
Whenever a user interacts with the app (e.g., enters text or clicks a button), the script reruns from top to bottom and updates the UI automatically.

---

## ⚡ Key Features

- Simple and fast development  
- No frontend knowledge required  
- Built-in UI components (text input, buttons, sliders, etc.)  
- Ideal for AI/ML and NLP applications  
- Easy to deploy and share  

---

## 🧩 Basic Example

```python
import streamlit as st

st.title("Hello Streamlit")
text = st.text_input("Enter something")

if text:
    st.write("You entered:", text)
